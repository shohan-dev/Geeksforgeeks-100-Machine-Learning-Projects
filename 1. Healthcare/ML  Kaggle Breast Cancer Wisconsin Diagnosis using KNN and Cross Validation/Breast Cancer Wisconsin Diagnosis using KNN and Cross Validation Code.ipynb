```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("./data.csv")
df
```

```python
df.isna().sum()
```

```python
df1 = df.drop(["id","Unnamed: 32"],axis=1)
df1["diagnosis"] = df1["diagnosis"].replace({"M": 1, "B": 0})
df1
```

```python
x = df1.drop("diagnosis",axis=1)
y = df1["diagnosis"]
```

```python
import seaborn as sns

sns.lmplot(x = 'radius_mean', y = 'texture_mean', hue = 'diagnosis', data = df1) 
```

```python
sns.lmplot(x ='smoothness_mean', y = 'compactness_mean', data = df1, hue = 'diagnosis') 
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 60)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors = 10) 
knn.fit(x_train, y_train)
knn.score(x_test,y_test) 

```

```python
neighbors = [] 
cv_scores = [] 

from sklearn.model_selection import cross_val_score 
# perform 10 fold cross validation 
for k in range(1, 50): 
	neighbors.append(k) 
	knn = KNeighborsClassifier(n_neighbors = k) 
	scores = cross_val_score( 
		knn, x_train, y_train, cv = 10, scoring = 'accuracy') 
	cv_scores.append(scores.mean()) 
```

```python
# get the error score
MSE = [1-x for x in cv_scores] 

# determining the best k 
optimal_k = neighbors[MSE.index(min(MSE))] 
print('The optimal number of neighbors is % d ' % optimal_k) 

# plot misclassification error versus k 
plt.figure(figsize = (10, 6)) 
plt.plot(neighbors, MSE) 
plt.xlabel('Number of neighbors') 
plt.ylabel('Misclassification Error') 
plt.show() 
```
